**Passo 3 - Corre√ß√£o do DataFrame (Limpeza, organiza√ß√£o e modelagem)**

---

```python
# Alterar as colunas para portugu√™s para facilitar a an√°lise para mim

traducao_colunas = ({
    "Transaction ID": "ID do produto",
    "Date": "Data",
    "Product Category": "Categoria",
    "Product Name": "Nome do produto",
    "Units Sold": "Quantidade",
    "Unit Price": "Preco unitario",
    "Total Revenue": "Preco total",
    "Region": "Regiao",
    "Payment Method": "Forma de pagamento"
    
})

df_vendas = df_vendas.rename(columns=traducao_colunas)

display(df_vendas.head())
```

```python
# Criar um novo dataframe para a nova organiza√ß√£o e modelagem da base de dados

# Excluir a coluna ID do produtro (essa coluna √© desnecess√°ria e n√£o influencia no EDA)
df_analise = df_vendas.drop(columns="ID do produto")

display(df_analise.head())
```

```python
# Alterar o formato da coluna Data de (yyyy-mm-dd) para o formato (month-yyyy)

# Carregar o dataframe
df_analise= pd.DataFrame(df_analise)

# Convertendo a coluna 'Data' para o tipo datetime
df_analise['Data'] = pd.to_datetime(df_analise['Data'])

# Criando a coluna Periodo com o formato 'M√™s/Ano'
df_analise['Periodo'] = df_analise['Data'].dt.strftime('%B %Y')

display(df_analise.head())
```

```python
# Excluir a coluna Data (N√£o √© mais necess√°ria)
df_analise = df_analise.drop(columns="Data")

display(df_analise.head())
```

```python
# Configura√ß√£o da ordem da coluna Periodo, que nos gr√°ficos s√£o exibidos fora de ordem. 

# Vari√°vel para ajustar o par√¢metro Categories nos gr√°ficos
ordem_categoria = ["January 2024", "February 2024", "March 2024", "April 2024", "May 2024", "June 2024", "July 2024", "August 2024"]

# Ordenando a coluna Periodo
df_analise["Periodo"] = pd.Categorical(df_analise["Periodo"], categories=ordem_categoria, ordered=True)
```

Dataframe corrigido e remodelado.

**Passo 4 - EDA - An√°lise de Dados e Visualiza√ß√£o de Dados**

---

```python
# EDA descritivo

# Agrupar pelo pre√ßo total, m√©dio, m√≠nimo, m√°ximo e quantidade de produtos vendidos
produto_categoria = df_analise.groupby(["Categoria", "Nome do produto"])["Preco total"].agg(
    ["sum", "mean", "min", "max", "count"])

# Configura√ß√£o das colunas
produto_categoria.columns = ["Pre√ßo Total", "Pre√ßo M√©dio", "Pre√ßo Min", "Pre√ßo Max", "Qtd. Vendas"]

# Visualizar os dados
display(produto_categoria)
```

```python
# An√°lise das Vendas e Faturamentos por Per√≠odo e Categoria

# An√°lise das vendas por per√≠odo e categoria
print("Vendas por categoria:")
pivot_quantidade = df_analise.pivot_table(values="Quantidade", index="Categoria", columns="Periodo", aggfunc="sum", 
                                          fill_value=0)

pivot_quantidade["Total Periodo"] = pivot_quantidade.sum(axis=1)

# Visualizar resultados
display(pivot_quantidade)

print(f"Total Geral: {pivot_quantidade["Total Periodo"].sum()} unidades\n")

# An√°lise dos faturamentos por per√≠odo e categoria
print("\nFaturamento das vendas por categoria:")
pivot_valor = df_analise.pivot_table(values="Preco total", index="Categoria", columns="Periodo", aggfunc="sum", 
                                     fill_value=0)

pivot_valor["Total Periodo"] = pivot_valor.sum(axis=1)

# Visualizar resultados
display(pivot_valor.round(2))

print(f"Valor Total Geral: US$ {pivot_valor['Total Periodo'].sum()}")

```

## **1. Analisar as tend√™ncias de vendas ao longo do tempo para identificar padr√µes sazonais ou oportunidades de crescimento.** ##

```python
# Agrupar per√≠odos por vendas
analise_periodo = df_analise.groupby("Periodo")["Quantidade"].sum().sort_values(ascending=False)

# Visualizar vendas por per√≠odo
display(analise_periodo)

# Em porcentagem %
total_periodo = analise_periodo.sum()
percentual_periodo = (analise_periodo / total_periodo * 100).round(2)
percentual_periodo_formatado = percentual_periodo.apply(lambda x: f'{x:.2f} %')

# Visualizar percentual das vendas por per√≠odo
display(percentual_periodo_formatado)


# Agrupar per√≠odo por faturamento
analise_valor = df_analise.groupby("Periodo")["Preco total"].sum().sort_values(ascending=False)

# Visualizar as vendas por faturamento
display(analise_valor)

# Em porcentagem %
total_valor = analise_valor.sum()
percentual_valor = (analise_valor / total_valor * 100).round(2)
percentual_valor_formatado = percentual_valor.apply(lambda x: f'{x:.2f} %')

# Visualizar percentual das vendas por valor
display(percentual_valor_formatado)


# Agrupar vendas por regi√£o
analise_regiao = df_analise.groupby(["Regiao", "Periodo"])[["Quantidade", "Preco total"]].sum()

# Visualizar vendas por regi√£o
print("Vendas nos Per√≠odos por Regi√£o:")

display(analise_regiao)

# Somar vendas por regi√£o
print("\nTotal das vendas nos Per√≠odos por Regi√£o:")
total_regiao = analise_regiao.groupby(["Regiao"]).sum()

# Visualizar total das vendas por regi√£o
display(total_regiao)

print(f"Total Geral das Vendas nos Per√≠odos por Regi√£o:\n{total_regiao.sum()}")

print(f"\nPercentual %:\n{((total_regiao / total_regiao.sum()) * 100).round(2)}")

```

## **3. Investigue o impacto dos m√©todos de pagamento no volume de vendas ou na receita.** ##

```python
# Agrupar as vendas por forma de pagamento
print("Vendas por forma de pagamento:")
analise_metodo_pagamento = df_analise.groupby(["Forma de pagamento"])[["Quantidade"]].count()

# Calcular a soma total da quantidade das vendas
soma_total_vendas = analise_metodo_pagamento["Quantidade"].sum()

# Calcular o percentual e exibir a coluna percentual % no dataframe
analise_metodo_pagamento["Percentual %"] = (analise_metodo_pagamento["Quantidade"] /soma_total_vendas) * 100

# Visualizar os resultados
display(analise_metodo_pagamento.round(2))

print(f"Quantidade Total: {soma_total_vendas}\nPercentual Total: {analise_metodo_pagamento["Percentual %"].sum():.2f} %")


# Agrupar as vendas por regi√£o e forma de pagamento 
print("\nVendas por forma de pagamento e regi√£o:")
analise_metodo_pagamento_regiao = df_analise.groupby(["Regiao", "Forma de pagamento"])[["Quantidade"]].count()

# Calcular o total geral da Quantidade
total_geral = analise_metodo_pagamento_regiao["Quantidade"].sum()

# Adicionar a coluna Percentual % ao dataframe
analise_metodo_pagamento_regiao["Percentual %"] = (analise_metodo_pagamento_regiao["Quantidade"] / total_geral) * 100

# Calcular o percentual % m√°ximo
total_percentual = analise_metodo_pagamento_regiao["Percentual %"].max()

# Calcular o percentual % geral
analise_metodo_pagamento_regiao["Percentual Geral %"] = analise_metodo_pagamento_regiao["Percentual %"] / total_percentual * 100

# # Visualizar os resultados no dataframe com a coluna de percentual geral
display(analise_metodo_pagamento_regiao.round(2))

print(f"Quantidade Total: {total_geral}\nPercentual Total: {analise_metodo_pagamento_regiao["Percentual %"].sum():.2f} %\nPercentual Geral Total: {analise_metodo_pagamento_regiao["Percentual Geral %"].sum()/3:.2f} %")

```
